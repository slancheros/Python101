```python
import gradio as gr
import matplotlib.pyplot as plt
import pandas as pd

def make_plot():
  ted = pd.read_csv('ted.csv')
  fig =  ted[ted.comments<1000].comments.plot(kind='hist')
  return fig.get_figure()

outputs = gr.Plot()

demo = gr.Interface(
    fn= make_plot,
    inputs=[],
    outputs=outputs
)

if __name__ == "__main__":
    demo.launch(debug="True")

```
